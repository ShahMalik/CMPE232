```python
class node():
    def __init__(self, name):
        self.name = name
        self.neighbors = []
        self.visited = False
        self.prev = None
        
class BreadthFirstSearch():
    def __init__(self):
        self.queue = []
    
    def bfs(self, s):
        self.queue.append(s)
        s.visited = True
        
        while len(self.queue) != 0:
            v = self.queue.pop(0)
            print("%s " % v.name);
            
            for neighbor in v.neighbors:
                if not neighbor.visited:
                    self.queue.append(neighbor)
                    neighbor.visited = True
                    neighbor.prev = v
        
```

```python
class node():
    def __init__(self, name):
        self.name = name
        self.neighbors = []
        self.visited = False
        self.prev = None
        
class DepthFirstSearch():
    def __init__(self):
        self.stack = []
    
    def dfs(self, s):
        self.stack.append(s)
        s.visited = True
        
        while len(self.stack) != 0:
            v = self.stack.pop()
            print("%s " % v.name);
            
            for neighbor in v.neighbors:
                if not neighbor.visited:
                    self.queue.append(neighbor)
                    neighbor.visited = True
                    neighbor.prev = v
        
```
